```python
for item in graph.edges:
    temp = graph.get_edge_data(item[0],item[1],item[2])
    print(item[0])
    print(item[1])
    print(temp['relation'])
    print()
```

```python
for item in graph.nodes:
    for item2 in graph.nodes:
        if graph.has_edge(item, item2):
            print(item, item2)
```

```python
for item in graph.get_edge_data:
    temp = graph.get_edge_data(item[0],item[1],item[2])
    print(item[0])
    print(item[1])
    print(temp['relation'])
    print()
```
