```python
##Label Encoder
## this is a part of data preprocessing where we are encoding the categorical variables

from sklearn.ensemble import RandomForestClassifier
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.feature_selection import SelectFromModel
from sklearn.metrics import accuracy_score
from sklearn import preprocessing


categorical_feats = [
    f for f in df.columns if df[f].dtype == 'object']

for col in categorical_feats:
    lb = preprocessing.LabelEncoder()
    lb.fit(list(df[col].values.astype('str')))
    df[col] = lb.transform(list(df[col].values.astype('str')))
        
## fixing missing value  
## the machine learning models don't run unless or until we get rid of missing values

df.fillna(-999, inplace = True)
```

```python
## splitting dataset into train, validate and test

train, validate, test = np.split(df.sample(frac=1), [int(.6*len(df)), int(.8*len(df))])
```

```python
y_train = train['total_time']
x_train = train.drop(['total_time'], axis=1)

y_validate = validate['total_time']
x_validate = validate.drop(['total_time'], axis=1)

y_test = test['total_time']
x_test = test.drop(['total_time'], axis=1)
```

<b>Here are the few machine learning models we tried training to get the prediction of "total time of commute"

```python
## LINEAR REGRESSION
from sklearn.linear_model import LinearRegression

reg = LinearRegression()

reg.fit(x_train, y_train)
```

```python
reg.score(x_validate,y_validate)
```

```python
## GRADIENT BOOSTING 
from sklearn import ensemble
clf = ensemble.GradientBoostingRegressor(n_estimators = 400, max_depth = 5, min_samples_split = 2,
                                        learning_rate = 0.1, loss = 'ls')
```

```python
clf.fit(x_train, y_train)
```

```python
clf.score(x_test, y_test)
```

```python
from sklearn import linear_model
```

```python
## RIDGE REGRESSION
Ridreg = linear_model.Ridge(alpha=.1)

Ridreg.fit(x_train, y_train)
```

```python
Ridreg.score(x_test, y_test)
```

```python
## LASSO REGRESSION

regr = linear_model.Lasso()
alphas = np.logspace(-4, -1, 6)
scores = [regr.set_params(alpha=alpha
          ).fit(x_train, y_train
             ).score(x_test, y_test)
       for alpha in alphas]
best_alpha = alphas[scores.index(max(scores))]
regr.alpha = best_alpha
regr.fit(x_train, y_train)
```

```python
## LOGISTIC REGRESSION
logistic = linear_model.LogisticRegression(C=1e5)
```

```python
logistic.fit(x_train, y_train)
```

```python
logistic.score(x_test, y_test)
```
